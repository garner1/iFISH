We consider the set of spotting probes locations as a 3D embedding of a topological shape - a graph - associated to the chromosome of interest. Why? Because...

* spotting probes are blind to their genomic localization, and in comparing different point clouds we have to do it independently of the probes' labelling: graphs spectral or intrinsic statistical properties are OK
* a graph representation is useful to describe the spatial structure of the point cloud, allowing the quantification of typical or rare spatial properties of the point cloud

The data is taken from the following path: /home/garner1/Work/dataset/fish/UserDots_IEG364_004

```python
import os
import csv
import numpy as np

xs = []
ys = []
zs = []
rows = 0

with open(os.path.join('/home/garner1/Work/dataset/fish/UserDots_IEG364_004', "014_001_2.csv"), "r")   as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        xs.append(float(row[0]))
        ys.append(float(row[1]))
        zs.append(float(row[2]))
        rows = rows + 1

M = np.empty([rows, 3])
ind = 0
with open(os.path.join('/home/garner1/Work/dataset/fish/UserDots_IEG364_004', "014_001_2.csv"), "r")   as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:   
        M[ind,0] = float(row[0])
        M[ind,1] = float(row[1])
        M[ind,2] = float(row[2])
        ind = ind + 1
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# ax.scatter(np.asarray(xs), np.asarray(ys),np.asarray(zs))
ax.scatter(xs, ys, zs)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
'''Create the distance matrix D'''

from scipy.spatial import distance_matrix

D = distance_matrix(M, M)

import seaborn as sns; sns.set()

# ax = sns.heatmap(D)
ax = sns.distplot(D.ravel())
plt.show()
```

```python
'''
Creat the graph from M
'''
import networkx as nx

pos3d = {i: (M[i,0], M[i,1], M[i,2]) for i in range(rows)}

r = 4000
G = nx.random_geometric_graph(rows, r, dim=3, pos=pos3d)

pos3d = nx.get_node_attributes(G, 'pos')
pos2d = {x:pos3d[x][:2] for x in pos3d}

nx.draw_networkx_edges(G, pos2d)
nx.draw_networkx_nodes(G, pos2d)
plt.show()
```

```python
'''Link graph and euclidean distances'''

p = nx.shortest_path(G)

gdist = []
edist = []
for k1,v1 in p.iteritems(): # the basic way
    for k2,v2 in v1.iteritems():
        gdist.append(len(v2)-1)
        edist.append(D[k1,k2])
```

```python
ax = sns.regplot(np.asarray(gdist), np.asarray(edist), fit_reg=False)
plt.show()
```
